# COVID-19 백신 종류별 접종 인구 

```python
days = [1, 2, 3]    # 1,2,3 Day
az = [2, 4, 8]      # { 단위 : 만명} 1일부터 3일까지 아스트라제네카 접종인구 수 
pfizer = [5, 1, 3]  # 5,1,3 Pfizer
moderna = [1, 2, 5] # 1,2,5 Moderna
```

```python
plt.plot(days, az)
plt.plot(days, pfizer)
plt.plot(days, moderna)
```

```python
plt.plot(days, az, label = 'az')    # label = 범례
plt.plot(days, pfizer, label = 'pfizer', marker='o', linestyle='--')    # marker = 마커모양, linestyle = 선모양
plt.plot(days, moderna, label='moderna', marker='s', ls='-.')   # ls = linestyle   

plt.legend()    # 범례를 표시
```

```python
plt.plot(days, az, label = 'az')    # label = 범례
plt.plot(days, pfizer, label = 'pfizer', marker='o', linestyle='--')    # marker = 마커모양, linestyle = 선모양
plt.plot(days, moderna, label='moderna', marker='s', ls='-.')   # ls = linestyle   

plt.legend(ncol=3)    # 범례를 표시
```
