# 04. 스타일 

```python
import matplotlib.pyplot as plt
import matplotlib
#matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 12   # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트시 minus 깨짐 현상을 해결함
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y, linewidth = 5)
```

```python
plt.plot(x, y, marker = 'o')
```

```python
plt.plot(x, y, marker='o', linestyle='None')
```

```python
plt.plot(x, y, marker='v', markersize=10)
```

```python
plt.plot(x, y, marker='D', markersize=15, markeredgecolor='red', markerfacecolor='yellow')
```

# 선 스타일 

```python
plt.plot(x, y, linestyle=':', color='red')
```

```python
plt.plot(x, y, linestyle='-.', color='red')
```

```python
plt.plot(x, y, linestyle='-', color='b')    #b:blue, g:green, r:red, c:cyan, m:magenta, y:yellow, k:black, w:white
```

## 포맷 

```python
plt.plot(x, y, 'ro--')  # color, marker, linestyle (r: red, o: circle, --: dashed line)
```

```python
plt.plot(x, y, 'bv:')   # color, marker, linestyle (b: blue, v: triangle_down, : dotted line)
```

```python
plt.plot(x, y, 'go')    
```

```python
plt.plot(x, y, color='g', marker='o', linestyle='None') #
```

# 설정 옵션 축약어 

```python
plt.plot(x, y, marker='o', mfc='red', mec='blue', ms=10, ls=':')    # mfc: marker face color, mec: marker edge color
```

# 선을 투명하게 

```python
plt.plot(x, y, marker='o', mfc='red', ms=10, alpha=0.3)   # alpha: 투명도
```

# 그래프 크기 

```python
plt.figure(figsize=(10, 5))
plt.plot(x,y)
```

```python
plt.figure(figsize=(5, 10))
plt.plot(x,y)
```

```python
plt.figure(figsize=(10, 5), dpi=150)    # dpi: 해상도 dot per inch  확대 비율
plt.plot(x, y)
```

# 배경색 

```python
plt.figure(facecolor='yellow')
plt.plot(x, y)
```

```python
plt.figure(facecolor='#a1c333', figsize=(10, 5))
plt.plot(x, y)
```
