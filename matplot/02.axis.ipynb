# 2. 축 Axis 

```python
import matplotlib.pyplot as plt
import matplotlib
#matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 12   # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트시 minus 깨짐 현상을 해결함
```

```python
x = [1, 2, 4]
y = [2, 4, 8]
```

```python
plt.plot(x, y)
#plt.title('꺾은 선 그래프', fontdic={'family':'HYGungSo-Bold', 'size':20})      #개별 폰트 사용 
plt.xlabel('X축', color='red', loc='right')  #  left, center, right 
plt.ylabel('Y축', color='#00aa00', loc='top')  #  top, center, bottom 
```

```python
plt.plot(x, y)
plt.xticks([1, 2, 3])  # x축 눈금
plt.yticks([2, 4, 6, 8])  # y축 눈금
plt.show()
```
