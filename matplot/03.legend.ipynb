# 03. 범례 (legend)

```python
import matplotlib.pyplot as plt
import matplotlib
#matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 12   # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트시 minus 깨짐 현상을 해결함
```

```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
plt.plot(x, y, label='무슨 데이터')
#plt.plot([1, 2, 3, 4],  [1, 4, 9, 16], label='무슨 데이터')
plt.legend(loc='lower right')
```

```python
plt.plot(x, y, label='범례')
plt.legend(loc='upper right')
```

```python
plt.plot(x, y, label='범례')
plt.legend(loc=(0.5, 0.5)) # x와 y의 0~1 사이의 값으로 위치 지정 
```
