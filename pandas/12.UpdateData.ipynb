# 12.데이터 수정 
column 수정

```python
import pandas as pd 
df = pd.read_csv('score.csv', index_col='지원번호')   #지원번호를 인덱스로 지정
df
```

```python
df['학교'].replace({'북산고':'상북고', '능남고':'무슨고'})     #북산고를 상북고로 변경
```

```python
df
```

```python
df['학교'].replace({'북산고':'상북고'}, inplace=True)     #북산고를 상북고로 변경]
df
```

```python
df['SW특기'].str.lower()
```

```python
df['SW특기'] = df['SW특기'].str.lower()     #SW특기를 소문자로 변경
df
```

```python
df['SW특기'] = df['SW특기'].str.upper()
df 
```

```python
df['학교'] = df['학교'] + '등학교'     #학교에 등학교 추가 
df
```

#column 추가 

```python
df['총합'] = df['국어'] + df['영어'] + df['수학'] + df['과학'] + df['사회']     #총합 추가
df
```

```python
df['결과'] = 'Fail'   #결과열 추가하고 Fail로 초기화
df
```

```python
df.loc[df['총합'] > 400, '결과'] = 'Pass'   #총합이 400점 이상이면 Pass로 변경
df
```

## column 삭제

```python
df.drop(columns=['총합'])    #총합 열 삭제
df
```

```python
df.drop(columns=['국어','수학','영어'], inplace=True)    #국어, 수학, 영어 열 삭제
df
```

```python
df.drop(index='4번')    #4번 학생 데이터를 삭제 
df
```

```python
filt = df['과학'] < 80  #수학 점수 80점 미만 학생 필터링
df[filt]
```

```python
df[filt].index
```

```python
df.drop(index=df[filt].index)
```
